## データの操作

### 準備：データベース、ファイルパスの設定

ノートブック名の右に`(SQL)`と表示されているのは、このノートブックのデフォルト言語がSQLであることを意味しています。クリックすることでデフォルト言語を変更することができます。ノートブックレベルでの言語指定に加えて、下のセルにあるようにセルレベルでの言語指定が可能です。`%python`というマジックワードを指定することで、当該セルの言語をPythonにすることができます。<br><br>

- [Databricksノートブックを使う \- Qiita](https://qiita.com/taka_yayoi/items/dfb53f63aed2fbd344fc#%E6%B7%B7%E6%88%90%E8%A8%80%E8%AA%9E)

DatabricksにはHiveメタストアが同梱されています。メタストアにデータを登録することで、SQLによるアクセスが可能になります。ファイルはDatabricks File System(DBFS)上に格納されます。DBFSの実態はS3です。<br><br>

- [Databricksにおけるデータベースおよびテーブル \- Qiita](https://qiita.com/taka_yayoi/items/e7f6982dfbee7fc84894)
- [Databricksにおけるファイルシステム \- Qiita](https://qiita.com/taka_yayoi/items/e16c7272a7feb5ec9a92)

```
%python
import re
from pyspark.sql.types import * 

# ログインIDからUsernameを取得
username_raw = dbutils.notebook.entry_point.getDbutils().notebook().getContext().tags().apply('user')
# Username の英数字以外を除去し、全て小文字化。Username をファイルパスやデータベース名の一部で使用可能にするため。
username = re.sub('[^A-Za-z0-9]+', '', username_raw).lower()

# データベース名
db_name = f"20210826_workshop_{username}"

# Hiveメタストアのデータベースの準備:データベースの作成
spark.sql(f"CREATE DATABASE IF NOT EXISTS {db_name}")
# Hiveメタストアのデータベースの選択
spark.sql(f"USE {db_name}")

# ファイル格納パス
work_path = f"dbfs:/tmp/20210826_workshop/{username}/"

print("database name: " + db_name)
print("path: " + work_path)
```

### 準備：作成済みテーブルの削除

以下のセルではデモで使用するテーブルの初期化を行なっています。ノートブックのデフォルト言語がSQLなので、以下のセルはSQLと解釈されます。

```
%sql
DROP TABLE IF EXISTS diamonds;
DROP TABLE IF EXISTS property_data_kanto;
DROP TABLE IF EXISTS property_kanto_delta;
DROP TABLE IF EXISTS property_summary_delta;
```

### Databricksデータセットからテーブルを作成する：CSVファイル

Databricks環境の`/databricks-datasets`配下には様々なサンプルデータが格納されています。定期的に更新されるCOVID-19データセットも含まれています。

[DatabricksにおけるCOVID\-19データセットの活用: データコミュニティで何ができるのか \- Qiita](https://qiita.com/taka_yayoi/items/3d62c4dbdc0e39e4772c)

```
%sql
-- テーブルが存在する場合には削除します
DROP TABLE IF EXISTS diamonds;

-- OPTIONSでcsvのファイルパス(path)、ヘッダーがあること(header)を指定し、USINGでファイルフォーマットを指定してdiamondsテーブを作成します
CREATE TABLE diamonds
USING csv
OPTIONS (path "/databricks-datasets/Rdatasets/data-001/csv/ggplot2/diamonds.csv", header "true")
```

```
%sql
-- 従来のデータベース同様にテーブルへの問い合わせを行うことができます
SELECT * from diamonds
```

### Blobからデータを取得する: Parquetファイル
<head>
  <link href="https://fonts.googleapis.com/css2?family=Kosugi+Maru&display=swap" rel="stylesheet">
  <style>
    * {font-family: "Kosugi Maru", sans-serif !important;}
  </style>
</head>
[Apache Parquetとは](https://databricks.com/jp/glossary/what-is-parquet)
> Parquetとは、Hadoopエコシステムの各種プロジェクトで利用可能なオープンソースのファイルフォーマットです。Apache Parquetは、CSVやTSVファイルのような行指向ファイル形式に対し、効率的で高性能な列指向ストレージ形式です。

ここではPythonを用いて関東地方の物件データを読み込みます。PySparkを使用してSparkにアクセスします。

- [Databricks Apache Sparkクイックスタート \- Qiita](https://qiita.com/taka_yayoi/items/bf5fb09a0108aa14770b)
- [Databricksにおけるデータのインポート、読み込み、変更 \- Qiita](https://qiita.com/taka_yayoi/items/4fa98b343a91b8eaa480)

```
%python

# 公開Azure Storage Blobから学習データを取得します (WASBプロトコル)
sourcePath = 'wasbs://mokmok@jixjiadatabricks.blob.core.windows.net/property_data_kanto/'

# PySpark APIを使用してSparkデータフレームを取得します
# format:読み取るファイルのフォーマットを指定します
# load:ファイルパス
# cache:データフレームをメモリーにキャッシュ
df = spark.read\
          .format('parquet')\
          .load(sourcePath)\
          .cache()   #イテレーティブに使うのでメモリーにキャッシュする

# テーブルとして登録 (Hiveマネージドテーブル)
df.write.mode('overwrite').saveAsTable('property_data_kanto')

# display関数を使用することで、データフレームの内容を並び替えたり、グラフで可視化を行うことができます。
display(df)
```

```
%python
# Sparkデータフレームでもpandasと同様のメソッド(一部)を利用できます。
df.count()
```

### Delta形式にする：高い信頼性と高性能なデータレイクへ

<img src="https://jixjiadatabricks.blob.core.windows.net/images/delta-lake-square-black.jpg" width="220">

**Delta Lakeの特徴**

1. ACIDトランザクションによるデータの信頼性確保
2. 大容量メタデータに対する高速分散処理
3. Time Travel (データのバージョン管理)
4. オープンフォーマット (Apache Parquet)
5. バッチとストリーミング両方に対応
6. スキーマ強制によるデータ品質の担保
7. 監査ログによるデータ捜査の追跡
8. データの更新・削除
9. Apache Spark API準拠

[Delta Lakeクイックスタートガイド \- Qiita](https://qiita.com/taka_yayoi/items/345f503d5f8177084f24)

```
%python
# デルタレイクにおけるフォルダを指定
DELTALAKE_BRONZE_PATH = f"{work_path}/delta/property_kanto"

# 既にフォルダがある場合は削除
dbutils.fs.rm(DELTALAKE_BRONZE_PATH, recurse=True)

# Delta形式で書き込みます。formatに"delta"を指定します。
df.write.format("delta").save(DELTALAKE_BRONZE_PATH)
```

### パスを指定して参照もできるが・・・

```
%python
# spark.sqlを使用することでSQLを記述できます
display(spark.sql("select * from delta.`{}` order by 2 desc limit 5".format(DELTALAKE_BRONZE_PATH)))
```

### パスを指定して参照もできるが、Deltaテーブル化した方がアクセスは簡単

```
%python
# SQL文でHiveメタストアにテーブルを作成します
display(spark.sql("DROP TABLE IF EXISTS property_kanto_delta;"))
display(spark.sql("CREATE TABLE property_kanto_delta USING DELTA LOCATION '{}';".format(DELTALAKE_BRONZE_PATH)))
```

```
%sql
-- describe detailでテーブルのメタデータを確認できます
describe detail property_kanto_delta
```

```
%sql
-- describe historyでテーブルの更新履歴を表示します
describe history property_kanto_delta
```

<br>
<h2>特徴量説明</h2>
<ul>
  <li>PropertyName: 物件名 or 建物名</li>
  <li>Price: 売却価格 (単位：万円)</li>
  <li>Prefecture: 物件所在都道府県</li>
  <li>Municipality: 物件市区町村</li>
  <li>Address: 物件住所 (プライバシー保護の為番地以降は省略)</li>
  <li>FloorPlan: 間取り (例：1LDK, 3LDK+S)</li>
  <li>PublicTransportName: 最寄り駅路線名 (例：JR中央線、都営三田線)</li>
  <li>NearestStation: 最寄り駅名</li>
  <li>DistanceMethod: 最寄り駅へのアクセス方法 (例：徒歩、バス、車)</li>
  <li>Distance_min: 最寄り駅へのアクセス時間 (単位：分)</li>
  <li>FloorSpace_m2: 専有面積 (単位：平米)</li>
  <li>OtherSpace_m2: バルコニー面積 (単位：平米)</li>
  <li>BuiltYear: 築年 (例：1998、2018)</li>
</ul>
<span style='color:green;'>
人工的に作られた特徴量：
<ul>
  <li>Age: 築年数 (BuiltYearから算出)</li>
  <li>FloorSpaceBucket: 専有面積をカテゴリ化 (10段階)</li>
  <li>AgeBucket: 築年数をカテゴリ化 (8段階)</li>
  <li>DistanceBucket: 最寄り駅アクセス分数をカテゴリ化 (21段階)</li>
</ul>
</span>
